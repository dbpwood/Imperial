### Project Overview

With an increase in the elderly population, home medical alert systems are a multi-million pound
and dollar business in Europe and the USA. These monitoring systems are often used by people with 
pre-existing medical conditions or post-operative patients who have returned home but are still 
required to be monitored in case of infection or other complications. The devices are usually worn 
by the user and send a signal to a specialized medical alert company that passes on any medical alerts 
from the devices to doctors and hospitals. 

There are many such devices out there in the marketplace and prices range from hundreds to tens of pounds 
or dollars depending on the type of monitoring being performed by the device. As such, competition in the 
market is strong, so companies must be competitive to sell their products. 

You are a research and development manager working in a small electronic company that manufactures a simple 
medical alert system called HeartSafe, which monitors users with heart conditions at home. If the user starts 
to display heart problems, such as Arrhythmia or fibrillation, the device signals an alert to the user and 
sends an emergency alert to a medical alert company. 

While this is a good selling and useful device, you believe that you can make a similar more enhanced 
device that could serve some members of the public better, while at the same time making more profit 
for the company. 

With your strong background in R&D and business management, you come up with the idea of a second device, 
which you call the SmartAlert system. This system also monitors respiration rate, blood oxygen, and 
blood pressure, so it can detect the early onset of a variety of medical issues including fainting, 
COPD, hypoxia as well as heart problems. 

After doing a technical and business evaluation of the idea you decide to do a business proposal 
presentation to the upper management in the company to try to convince them of the merits of producing 
this new product. In your business proposal, you will have to show if it is more profitable to just 
replace the present HeartSafe device with the new SmartAlert device or if you should produce some 
combination of the two different devices. 

For simplicity, we will refer to these device types as Type X for the HeartSafe device and Type Y 
for the SmartAlert device. 

With the machinery you have right now, you see that you could make 200 Type X devices per hour, 
but you could also make 140 Type Y devices per hour on the same machinery. After some calculations, 
you note that the profit per Type X device is \\$25 and the profit per Type Y device is \\$30. 
You have also discovered that in a week you can create a maximum of 6000 Type X devices and 4000 Type Y 
devices because of the person-hours and capacity of specialists available to you. 

Note that you have only <u>**40 hours per week**</u> to produce them.

## Problem Statement
 <br>

**$x_0$ = Type X, HeartSafe system** <br>
**$x_1$ = Type Y, SmartAlert system** <br> <br>


**Objective Function**  <br><br> 

$Profit = 25 x_0 + 30 x_1$ <br> <br>

***Constraints***

* Machinery constraint  

    $0.7 x_0 + 1.0 x_1 \leq 5600$  <br> <br>(This relates to the machinery and hours per week capacity constraint - I have factored up to a more acceptable and visually manageable function, as opposed to $1/200 x_0 + 1/140 x_1 \leq 40 $) <br> <br>

* Type X capacity constraint   

    $x_0 \leq 6000$  <br><br>

* Type Y capacity constraint
    
    $x_1 \leq 4000$ <br><br>

* Type Y Production constraint 

    $x_0 , x_1 \geq 0$  <br><br>
  
We choose the initial condition,  $i_0$, to be origin: $i_0 = [0.0,0.0]$ <br><br>

* Calculate the final profit of the production of devices
    
* What would happen if your product presents any defects?
    
* How to compete in a market that is new for you

## Solving the problem visually

### Defining $x_0$ and $x_1$

`np.linspace` is a function of `numpy` that allows us to get equally spaced values between a given range

```python
x_0 = np.linspace(0, 8000, 100000) # generate 6000 equaly spaced values between 0 and 6000 
```

We will make use of $x_0$ and the equation to define our lines and plot them. 
You can define the lines in terms of $x_1$ too.

The second step is to generate these lines on the graph. That is the first step in solving the problem

```python
# determine the plot size
plt.figure(figsize= (16, 10))

# plotting the lines
plt.plot(x_0, 5600 - 0.7*(x_0), color = 'y', linewidth=4, label = '$0.7x_0 + 1.0x_1 \leq 5600$ : Machinery constraint')  # constraint 1
plt.axvline(x = 6000, color = 'b', linewidth=4, label = '$x_0 \leq 6000$ : Type X capacity')  # constraint 2
plt.axhline(y = 4000, color = 'r', linewidth=4, label = '$x_1 \leq 4000$ : Type Y capacity')  # constraint 3

# limits
plt.xlim((0, 8500))
plt.ylim((0, 6000))
plt.xlabel('Item ($x_0$)', fontsize = 20)
plt.ylabel('Item ($x_1$)', fontsize = 20)

# legend
plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)


plt.show();
```

### Feasible Region

The area which takes both the constraints into consideration is called the feasible region

```python
# determine the plot size
plt.figure(figsize= (16, 10))

# plotting the lines
plt.plot(x_0, 5600 - 0.7*(x_0), color = 'y', linewidth=4, label = '$0.7x_0 + 1.0x_1 \leq 5600$ : Machinery constraint')  # constraint 1
plt.axvline(x = 6000, color = 'b', linewidth=4, label = '$x_0 \leq 6000$ : Type X capacity')  # constraint 2
plt.axhline(y = 4000, color = 'r', linewidth=4, label = '$x_1 \leq 4000$ : Type Y capacity')  # constraint 3

# limits, titles and labels
plt.xlim((0, 8500))
plt.ylim((0, 6000))
plt.xlabel('Item X or ($x_0$)', fontsize = 20)
plt.ylabel('Item Y or ($x_1$)', fontsize = 20)
plt.title('A Feasibility Graph to Visualise Our Constraints', fontsize=30)

# fill in the feasible region
plt.fill_between(x_0, # regular item 1 
                 np.minimum(5600 - 0.7*(x_0), 4000), # choose the minimum between Y from two eqns
                 where = x_0 <= 6000 , # defines any condition that might need to be considered
                 color ='orange', # shaded region
                 alpha = 0.5) # transparency level

# adjusting legend placement
plt.legend(bbox_to_anchor=(0.65, 0.95), fontsize=15, loc=1, borderaxespad=0.)

# export the file
plt.savefig("linear_opt", dpi = 300)

# plot show
plt.show();
```

The solution should be a combination of $x_0$ and $x_1$ that should be in the feasible region and maximises the profit function that we have $25 x_0 + 30 x_1$


## Solving the Problem using Python

[Scipy Linear Prog](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.linprog.html)

We will be solving the problem using the `scipy.optimize` library and `linprog`

### Using a simple approach using Scipy Linear Prog

```python
from scipy.optimize import linprog
```

```python
# Construct parameters
profit = np.array([25, 30])

# Inequality constraints
capacity = np.array([[1, 0],
                      [0, 1], 
                         [0.7, 1]])

constraints = np.array([6000, 4000, 5600])

# Solve the problem
# we put a negative sign on the objective as linprog does minimization by default
production = linprog(-profit, A_ub = capacity, b_ub = constraints, method='revised simplex')

production
```

```python
data = {
    'Optimal result':[6000, 1400], 
    'Slack': [0, 2600],
}

df_results = pd.DataFrame(data, index=['X', 'Y'])

# Create a stacked bar plot
df_results.plot(kind='bar', stacked=True)

plt.title('Optimal Number of Devices to Maximise Profit')

# export the file
plt.savefig("results", dpi = 300)

plt.show()
```

### Using a step by step approach using Scipy minimize

Getting a maximise result by setting the `sign` perameter to `-1`

```python
from scipy.optimize import minimize
```

**Defining the objective function**

We will begin by defining the objective function

$25x_0 + 30x_1$

This can be done by defining a function, in this case called `objective_fun`, that contains the definition of the function we want to maximise. Note that this function takes one argument, `x`, that is an array containing our DOF `x[0]`,`x[1]`.

```python
def objective_fun(x, sign=-1.0):
    x0 = x[0]
    x1 = x[1]
    return sign*(25*x[0] + 30*x[1])
```

**Defining the constraints**

`SciPy optimize` can handle both linear and non-linear constraints, with equalities or inequalities. We need to be little careful with how we define the constrains in Python. In fact, a constraint function $f(x)$ needs to be written as 

$$f(x) = 0,$$

in case of an equality, or as 

$$f(x) \geq 0,$$

in case of an inequality. Therefore, in some cases we might need to manipulate the equations.

Additionally, each constraint needs to be defined in a separate function. For this problem, we will only have to define 1 constraint function.

```python
def constraint1(x):
    return 5600 - 0.7*x[0] - x[1]
```

We are alomost done with the constraints. The last thing we need to do is to tell Python the type of each constraint: equality `eq` or inequality `ineq`.

```python
con1 = {"type": "ineq", "fun" : constraint1}
```

FInally, we collect all these dictonaries inside a list, `cons`, for future use.

```python
cons = [con1]
```

**Definition of the bounds**

In this step, we will define the bounds for each of the DOF as defined at the beginning of this notebook.

In the code cell below, we have defined in a tuple `b_0` the lower and upper bounds for `x_0`. Then in a similar way we have defined in a tuple `b_1` the lower and upper bounds for `x_1`.

```python
b_0 = (0.0, 6000.0)
b_1 = (0.0, 4000.0)
```

Finally, we collect all these tuples as an array, `bnds`, for future use.

```python
bnds = (b_0, b_1)
```

**Definition of the initial guess**

We run the code cell below to define the initial guess `i_0`.

```python
i_0 = [0,0]
```

**Solving the optimization problem**

Now we are ready to solve our problem. As mentioned earlier, we will use the function `SciPy optimize` `linprog` with the following arguments:

- `objective_fun`: the name of the function where we have defined the fuction that we want to maximise.
- `i_0`: the initial condition.
- `method = ‚Äôsimplex‚Äô`: the optimization algorith chosen to solve the problem. You can find a comprehensive list of the options available [here](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html).
- `bounds = bnds`: the bounds for our problem.
- `constraints = cons`: the constraints of our problem.

By running the code cell below to solve our problem!

```python
sol = minimize(objective_fun,  i_0, method='SLSQP', bounds = bnds, constraints = cons)
```

```python
print(sol)
```

We observe that `sol` is an object representing the optimization results. This description, among other things, includes a feedback about wheter the algorith ran successfully or not, the minimization function (jac) and so on.

The solution vector `x` is also part of these attributes. Notice that it contains the optimal solution found by the solver and that (as expected!) matches the one we found by solving the same exercise in `linprog` and Excel.

For a complete description of the attributes returned please see [here](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.OptimizeResult.html#scipy.optimize.OptimizeResult).
